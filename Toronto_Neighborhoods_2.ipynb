```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
#print(soup.prettify())
table = soup.find('table')
```

```python
table
```

```python
df = pd.read_html(str(table))
```

```python
dataframe = pd.DataFrame(df[0])
```

```python
dataframe.columns = dataframe.iloc[0]
```

```python
dataframe.drop(dataframe.index[0], inplace = True)
```

```python
dataframe['Borough'].replace('Not assigned', np.nan, inplace = True)
```

```python
dataframe.dropna(subset = ['Borough'], axis = 0, inplace = True)
```

## Showing Neighborhoods

```python
import folium 
latitude_toronto = 43.651070
longitude_toronto = -79.347015
```

```python
map_Toronto = folium.Map( location=[latitude_toronto, longitude_toronto], zoom_start=10 )

# add markers to map
for lat, lng, borough, neighborhood in zip( data_frame['Latitude'], data_frame['Longitude'], data_frame['Borough'], data_frame['Neighbourhood'] ):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.8,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
